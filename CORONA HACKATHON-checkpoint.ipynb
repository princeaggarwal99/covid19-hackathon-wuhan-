# Model-1 ( taking most recent visit of patient to hospital and filling missing values with -1)

```python
# TAKING THE  ROW CORRESPONDING TO THE MOST RECENT VISIT OF PATIENT TO THE HOSPITAL :( ALSO REPLACING NAN WITH "-1")
```

```python
# uploading the train dataset and filling the patient id for nan vaalues
data=pd.read_excel('train.xlsx')
data['PATIENT_ID'].fillna(method='ffill',inplace = True)
model1=data.drop_duplicates(subset='PATIENT_ID',keep='last') # removing all ROWS except the most recent visit to the hospiital
```

```python
model1.head(10)
```

```python
model1.isnull().sum() #NULL VALUES
```

```python
model1['RE_DATE'].fillna(method='ffill',inplace=True) # FILLING THE REDATE 
```

```python
model1.fillna('-1',inplace=True)  ## REPLACING NULL VALUES WITH (-1)
```

```python
model1.head()
```

```python
model1.tail()
```

```python
model1.dtypes
```

```python
model1=model1.drop(['RE_DATE','Admission time','Discharge time'],axis=1) # REMOVING COLUMNS NOT MUCH NECESSARY FOR MODEL BUILDING
```

```python
model1.head()
```

```python
#SEPERATING TARGET AND FEATURES
y=model1['outcome']
x=model1.drop('outcome',axis=1)
```

```python
x
```

```python
y
```

```python
## importing METRICES , other models 
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score,GridSearchCV
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier,GradientBoostingClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=.25)
```

```python
# HYPER-PARAMETER TUNING : FINDING THE BEST PARAMETERS FOR THE MODEL
```

```python
## SVC SUPPORT VECTOR CLASSIFIER 
from sklearn.svm import SVC
svc=SVC()
param = {
    'C': [0.1,0.8,0.9,1,1.1,1.2,1.3,1.4],
    'kernel':['linear', 'rbf'],
    'gamma' :[0.1,0.8,0.9,1,1.1,1.2,1.3,1.4]
}
grid_svc = GridSearchCV(svc, param_grid=param)
grid_svc.fit(x_train, y_train)
grid_svc.best_params_
```

```python
# RANDOM FOREST CLASSIFIER
rfc=RandomForestClassifier()
parameters={'n_estimators':[1,10,50,100,200,300,400,500,1000]}
clfr=GridSearchCV(rfc,parameters)
clfr.fit(x,y)
clfr.best_params_
```

```python
## KNEIGHBOURS CALSSIFIER ()
knn=KNeighborsClassifier()
parameters={'n_neighbors':[i+1 for i in range(10)]}
clf=GridSearchCV(knn,parameters)
clf.fit(x,y)
clf.best_params_
```

```python
# gradient boosting CLASSIFIER
gbc=GradientBoostingClassifier()
parameters={'n_estimators':[1,10,50,100,200,300,400,500,1000],'learning_rate':[0.001,0.01,0.1,1]}
clfr=GridSearchCV(gbc,parameters)
clfr.fit(x,y)
clfr.best_params_
```

```python
# ADA boost  CLASSIFIER
ada=AdaBoostClassifier()
parameters={'n_estimators':[1,10,50,100,200,300,400,500,1000],'learning_rate':[0.001,0.01,0.1,1]}
clfr=GridSearchCV(ada,parameters)
clfr.fit(x,y)
clfr.best_params_
```

```python
# decision tree classifier
dtc=DecisionTreeClassifier()
parameters={'criterion':['gini','entropy']}
clfr=GridSearchCV(dtc,parameters)
clfr.fit(x,y)
clfr.best_params_
```

```python
## tuning the parameters
rfc=RandomForestClassifier(n_estimators =1)
knn=KNeighborsClassifier(n_neighbors=1)
svc=SVC(gamma=0.1,kernel='linear')
ada=AdaBoostClassifier(n_estimators =1,learning_rate=0.001)
gbc=GradientBoostingClassifier(n_estimators =100,learning_rate=0.001)
dtc=DecisionTreeClassifier(criterion='gini')
LR=LogisticRegression()
GNB=GaussianNB()
```

```python
models = []
models.append(('randomforestclassifier', rfc))
models.append(('kneighborsclassifier', knn))
models.append(('svc', svc))
models.append(('ada boost classifier', ada))
models.append(('gradient boosting classifier', gbc))
models.append(('decisiontreeclassifier', dtc))
models.append(('LOGISTIC REG', LR))
models.append(('gaussian nb', GNB))

```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix,auc,roc_auc_score,roc_curve,f1_score
from sklearn.model_selection import cross_val_score
```

```python
Model=[]
score=[]
cvs=[]
rocscore=[]
F1=[]
for name,model in models:
    print('********',name,'*******')
    print('\n')
    Model.append(name)
    model.fit(x_train,y_train)
    print(model)
    pre=model.predict(x_test)
    print('\n')
    AS=accuracy_score(y_test,pre)
    print("acc_score=",AS)
    score.append(AS*100)
    print('\n')
    f1=f1_score(y_test,pre)
    print("F1_score=",f1)
    F1.append(f1*100)
    print('\n')
    sc=cross_val_score(model,x,y,cv=10,scoring='accuracy').mean()
    print('cross_val_sc=',sc)
    cvs.append(sc*100)
    print('\n')
    fpr,tpr,thresholds=roc_curve(y_test,pre)
    roc_auc=auc(fpr,tpr)
    print("roc_auc_sc=",roc_auc)
    rocscore.append(roc_auc*100)
    print('\n')
    print('classification_repo\n',classification_report(y_test,pre))
    print('\n')
    cm=confusion_matrix(y_test,pre)
    print(cm)
    print('\n')
    plt.figure(figsize=(10,40))
    plt.subplot(911)
    plt.title(name)
    print(sns.heatmap(cm,annot=True))
    plt.subplot(912)
    plt.title(name)
    plt.plot(fpr,tpr)
    plt.plot([0,1],[0,1],'r--')
    plt.xlabel('TPR')
    plt.ylabel('FPR')
    print('\n\n')
```

```python
result=pd.DataFrame({'model':Model,'Accu_score':score,'f1_score':F1,'cross_val_Score':cvs,'roc-score':rocscore})
result
```

```python
# SUPPORT VECTOR MACHINE IS WORKING BEST AS THE ACCURACY SCORE IS 100 THOUGH THE CROSS VALITDATION SCORE IS ALSO GOOD ALMOST 500 
```
